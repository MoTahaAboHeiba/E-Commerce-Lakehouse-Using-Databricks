#The Transformation Logic

```python
query ='''
select  
    ROW_NUMBER() OVER (ORDER BY p.start_date, p.product_number) AS product_key,
    p.product_id,
    p.product_number,
    c.category_id,
    p.product_name,
    c.category,
    c.subcategory,
    p.product_cost,
    p.product_line,
    c.maintenance_flag,
    p.start_date,
    p.end_date

from `e-commerce-databricks-project`.silver_layer.crm_product_info p
    left join `e-commerce-databricks-project`.silver_layer.erp_categories_details c
        on p.category_id = c.category_id
'''
df = spark.sql(query)
```

```python
display(df)
```

#Writing Gold Table

```python
df.write.mode("overwrite").format("delta").saveAsTable("`e-commerce-databricks-project`.gold_layer.Dim_products")
```
